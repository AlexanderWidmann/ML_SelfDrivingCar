# Project Proposal for Artificial Intelligence & Machine Learning
The goal of this project is to learn new methods for machine learning and also use them in praxis.

## Project description
The goal of the project is to create a self-driving car.
To implement this the Udacity car simulator (https://github.com/udacity/self-driving-car-sim) is used.<br>
During this project multiple different models should be implemented and tested. Since the parameters can have a pretty pig impact on the results of the model it is also important to observe the models with different parameters to get their best results.

## Motivation
The reason why i chose this specific project is relatively simple. I am interested in the future and possiblities of completly self-sufficient cars and how that could impact our daily lives.With the udacity car simluator it is possible for me to actually test some algorithms and methods for myself.<br>

## Goal
Create working implementation of different models and test how well they do for different tracks.<br>
The models will only generate steering angles from the input pictures in the first iteration.<br>
The tracks should incude at least:<br>
 - Track, that generated the data
 - One different track
Bonus objecive:
 - One similar track
 - One different track
Example: snow vs. grass on the side of the road
Observations:
 - Can the model finish the track?
 - Determine and analyse problem areas
 - How does the model handle new tracks?
 
### Bonus goal
To keep the project small acceleration and breaking will not be considered for the first implementation. This will also be kept in mind for the data generation. If the models work well and the project progresses as planned I will also implement the speed part of the self driving car. This will also create more interesting reinforcement lerning task since the car should go as fast as possible to get the maximum rewards.

## Data generation
The simulator already provides a tool to generate data for the models to use. To keep the processing time small every produced dataset should be between 1 and 5 laps for any given track.
It is also possible to combine different datasets to get more training data.

## Approach
The approach is split into different steps that make the project easy to expand and still keep each step simple and replicable.

### Step 1: Proof of concept
This step is actually not directly part of the project but is rather a step to ensure the wanted outcome can be achieved. In this step I want to show, that it is in fact possible to create an machine learning algorithm that creates a self-driven car for the udacity car simluator. To accomplish this step I will use a working example from github to implement it and use as a basis for further development.<br>
I decided to use the following project: https://github.com/llSourcell/How_to_simulate_a_self_driving_car

### Step 2: First model creation and tests
Using the prototype created with Step 1 I will start to adjust parameters, add new parameters and try to tweak the algorithm to observe the effects and find a good implementation of the Sequential Model. This model is a relatively simple model that just maps images of the traning data to the given outputs and then correlates similiar inputs accordingly.<br>
While this is model will probably struggle in completly new environments it is still worthwhile to observe.

### Step 3: Find, implement and test at least one other non-reeinforment learning model
Using only one model from start to finish can limit the success of a project so in this step i plan to read more literature to find at least another model i can use to generate a mapping from the input data.

## Step 4: Reeinforment learning
In this step I will create a model that uses reeinforcement learning to keep the car on track.<br>
This model should be able to start learning without the need for any input data. <br>
An important part of this is how the reward is defined. At the start I will go with a very simple process of rewarding a small amount for distance travelled and subtracting a big amount for going of track. After creating a succesfull model it is again important to test if a different reward function can result in a better trained model. <br>
This step will most likely be based on the Markov Decission Process. This is subject to change, if a better process is found.
## Risk management
### Simulator issues
In my research I found that the chosen simluator may not support reeinforcement learning or may not be a good solution for all the steps.
Therefore there are options for using two other simulators:<br>
 - deepdrive: https://github.com/deepdrive/deepdrive <br>
 - carla: http://carla.org/
The currently selected simulator is generally subject to change until the testing phase, since the models should be usable by both simuators and swapping may provide benefits that outweigh the time spent to incorporate the new simulator.


### Project failure
If for unknow reason I can not get a working prototype for most of the steps in a reasonable time or some additional reason is revealed why the project will be unable to conclude by the deadline (Ex.: Training time is too high) i will go on the following backup project: <br>
Using OpenAi Gym (https://gym.openai.com/) implement a reinforcement learning model for an atari game. This will again be split into 2 different categories: RAM-based or Image-based. This implementation would start with the RAM-based variant. The goal for the model here would be to score the highest possible amount of points.
